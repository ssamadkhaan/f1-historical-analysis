```python
import fastf1
import pandas as pd
import matplotlib.pyplot as plt

fastf1.Cache.enable_cache("../cache")
```

```python
races = [
    "Bahrain",
    "Saudi Arabia",
    "Australia",
    "Japan",
    "Monaco"
]
```

```python
winners = []

for race_name in races:
    session = fastf1.get_session(2024, race_name, "R")
    session.load()

    winner = session.results.iloc[0]

    winners.append({
        "Race": race_name,
        "Driver": winner["FullName"]
    })
```

```python
winners_df = pd.DataFrame(winners)

winners_df
```

```python
win_counts = winners_df["Driver"].value_counts()

win_counts
```

```python
plt.figure(figsize=(8,5))

win_counts.plot(kind="bar")

plt.title("Race Wins Across Selected 2024 Grands Prix")
plt.xlabel("Driver")
plt.ylabel("Wins")

plt.show()
```
