# Formula 1 Historical Analysis (1950–2024)

This project explores Formula 1 history using Pandas and Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt

drivers = pd.read_csv("../data/drivers.csv")
results = pd.read_csv("../data/results.csv")
constructors = pd.read_csv("../data/constructors.csv")
races = pd.read_csv("../data/races.csv")
```

## Question 1: Who are the Top 10 Drivers by Race Wins?

```python
winners = results[results["positionOrder"] == 1]

winner_names = winners.merge(
    drivers,
    on="driverId"
)

top10_drivers = (
    winner_names["surname"]
    .value_counts()
    .head(10)
)

top10_drivers
```

```python
plt.figure(figsize=(10,5))

top10_drivers.plot(kind="bar")

plt.title("Top 10 Drivers by Race Wins")
plt.xlabel("Driver")
plt.ylabel("Wins")

plt.xticks(rotation=45)

plt.show()
```

### Key Insight

Lewis Hamilton is the most successful driver in Formula 1 history with 105 race wins, followed by Michael Schumacher with 97 wins.

Max Verstappen is already third despite still being active.

## Question 2: Which Constructors Have the Most Race Wins?

```python
team_wins = winners.merge(
    constructors,
    on="constructorId"
)

top10_teams = (
    team_wins["name"]
    .value_counts()
    .head(10)
)

top10_teams
```

```python
plt.figure(figsize=(10,5))

top10_teams.plot(kind="bar")

plt.title("Top 10 Constructors by Race Wins")
plt.xlabel("Constructor")
plt.ylabel("Wins")

plt.xticks(rotation=45)

plt.show()
```

### Key Insight

Ferrari is the most successful constructor in Formula 1 history with 249 race wins.

McLaren ranks second with 185 wins, while Mercedes and Red Bull have achieved remarkable success in the modern era.

Williams remains one of the most successful teams historically despite not being a front-runner in recent seasons.

## Question 3: Which Countries Have Produced the Most Formula 1 Drivers?

```python
top10_nationalities = (
    drivers.groupby("nationality")
    .size()
    .sort_values(ascending=False)
    .head(10)
)

top10_nationalities
```

```python
plt.figure(figsize=(10,5))

top10_nationalities.plot(kind="bar")

plt.title("Top 10 Nationalities in Formula 1")
plt.xlabel("Nationality")
plt.ylabel("Number of Drivers")

plt.xticks(rotation=45)

plt.show()
```

### Key Insight

British drivers dominate Formula 1 history with 166 drivers in the dataset.

The United States ranks second with 158 drivers, reflecting the large number of American entries during Formula 1's early decades.

Italy, France, and Germany have also contributed significantly to the sport's history.

## Question 4: How Has the Number of Formula 1 Races Changed Over Time?

```python
races["decade"] = (races["year"] // 10) * 10

races_per_decade = (
    races.groupby("decade")
    .size()
    .sort_index()
)

races_per_decade
```

```python
plt.figure(figsize=(10,5))

races_per_decade.plot(marker="o")

plt.title("Number of Formula 1 Races by Decade")
plt.xlabel("Decade")
plt.ylabel("Number of Races")

plt.grid(True)

plt.show()
```

### Key Insight

The number of Formula 1 races has steadily increased over time.

The 2010s recorded the highest number of races (198), while the 2020s are on track to surpass that figure as Formula 1 continues to expand its global calendar.
