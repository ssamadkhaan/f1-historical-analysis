```python
results = pd.read_csv("../data/results.csv")
```

```python
results.head()
```

```python
results.shape
```

```python
results.columns
```

```python
winners = results[results["positionOrder"] == 1]
```

```python
winners.head()
```

```python
winners.shape
```

```python
winner_names = winners.merge(
    drivers,
    on="driverId"
)
```

```python
winner_names.head()
```

```python
winner_names["surname"].value_counts()
```

```python
winner_names["surname"].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
top10 = winner_names["surname"].value_counts().head(10)

top10.plot(kind="bar")

plt.title("Top 10 Drivers by Race Wins")
plt.xlabel("Driver")
plt.ylabel("Wins")

plt.show()
```

```python
plt.figure(figsize=(10,5))

top10.plot(kind="bar")

plt.title("Top 10 Drivers by Race Wins")
plt.xlabel("Driver")
plt.ylabel("Wins")

plt.xticks(rotation=45)

plt.show()
```

```python
constructors = pd.read_csv("../data/constructors.csv")
```

```python
constructors.head()
```

```python
constructors.columns
```

```python
team_wins = winners.merge(
    constructors,
    on="constructorId"
)
```

```python
team_wins.head()
```

```python
team_wins["name"].value_counts()
```

```python
top_teams = team_wins["name"].value_counts().head(10)

top_teams
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

top_teams.plot(kind="bar")

plt.title("Top 10 Constructors by Race Wins")
plt.xlabel("Constructor")
plt.ylabel("Wins")

plt.xticks(rotation=45)

plt.show()
```

```python
races = pd.read_csv("../data/races.csv")
```

```python
races.head()
```

```python
races.columns
```

```python
races["decade"] = (races["year"] // 10) * 10
```

```python
races.groupby("decade").size()
```

```python
races.groupby("decade").size().sort_index()
```

```python
races_per_decade = races.groupby("decade").size().sort_index()

plt.figure(figsize=(10,5))

races_per_decade.plot(marker="o")

plt.title("Number of F1 Races by Decade")
plt.xlabel("Decade")
plt.ylabel("Number of Races")

plt.grid(True)

plt.show()
```
