```python
import fastf1
import pandas as pd
import matplotlib.pyplot as plt

fastf1.Cache.enable_cache("../cache")

session = fastf1.get_session(2024, "Monaco", "Q")
session.load()
```

```python
session.results.head()
```

```python
top10 = session.results[["Abbreviation", "Q3"]].head(10)

top10
```

```python
q3_seconds = top10["Q3"].dt.total_seconds()

q3_seconds
```

```python
plt.figure(figsize=(10,5))

plt.bar(top10["Abbreviation"], q3_seconds)

plt.title("Monaco 2024 Qualifying Top 10")
plt.xlabel("Driver")
plt.ylabel("Lap Time (seconds)")

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.bar(top10["Abbreviation"], q3_seconds)

plt.title("Monaco 2024 Qualifying Top 10")
plt.xlabel("Driver")
plt.ylabel("Lap Time (seconds)")

plt.ylim(70,72)

plt.show()
```

# Conclusion

Charles Leclerc achieved pole position at Monaco 2024.

The top 10 drivers were separated by less than one second,
showing how competitive modern Formula 1 qualifying is.
