```python
import fastf1
import pandas as pd
import matplotlib.pyplot as plt

fastf1.Cache.enable_cache("../cache")
```

```python
race = fastf1.get_session(2024, "Monaco", "R")
race.load()
```

```python
race.results.head()
```

```python
top10 = race.results[["Abbreviation", "Position"]].head(10)

top10
```

```python
plt.figure(figsize=(10,5))

plt.bar(top10["Abbreviation"],
        top10["Position"])

plt.title("Monaco 2024 Race Results")
plt.xlabel("Driver")
plt.ylabel("Finishing Position")

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.bar(top10["Abbreviation"],
        top10["Position"])

plt.gca().invert_yaxis()

plt.title("Monaco 2024 Race Results")
plt.xlabel("Driver")
plt.ylabel("Finishing Position")

plt.show()
```
