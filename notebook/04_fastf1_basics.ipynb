```python
import matplotlib.pyplot as plt

top10 = session.results[["Abbreviation", "Q3"]].head(10)

q3_seconds = top10["Q3"].dt.total_seconds()

plt.figure(figsize=(10,5))
plt.bar(top10["Abbreviation"], q3_seconds)

plt.title("Monaco 2024 Qualifying Top 10")
plt.xlabel("Driver")
plt.ylabel("Lap Time (seconds)")

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.bar(top10["Abbreviation"], q3_seconds)

plt.title("Monaco 2024 Qualifying Top 10")
plt.xlabel("Driver")
plt.ylabel("Lap Time (seconds)")

plt.ylim(70, 71.5)

plt.show()
```

```python
lec_lap = session.laps.pick_driver("LEC").pick_fastest()
ver_lap = session.laps.pick_driver("VER").pick_fastest()

print("Leclerc:", lec_lap["LapTime"])
print("Verstappen:", ver_lap["LapTime"])
```

```python
lec_tel = lec_lap.get_car_data()
ver_tel = ver_lap.get_car_data()

lec_tel.head()
```

```python
lec_tel.columns
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

plt.plot(
    lec_tel["Time"].dt.total_seconds(),
    lec_tel["Speed"],
    label="Leclerc"
)

plt.plot(
    ver_tel["Time"].dt.total_seconds(),
    ver_tel["Speed"],
    label="Verstappen"
)

plt.title("Monaco 2024 Qualifying Speed Comparison")
plt.xlabel("Time (seconds)")
plt.ylabel("Speed (km/h)")
plt.legend()

plt.show()
```
